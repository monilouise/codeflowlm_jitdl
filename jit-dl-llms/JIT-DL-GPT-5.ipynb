```python
prompt_preffix = """
You are a static analysis and fault localization expert.  

The following GitHub diff introduces a bug. Please analyze specifically the changed lines in the diff for potential bugs.  DO NOT consider the commented lines.  

Return the results in JSON format, consisting of a single JSON object with two fields: “intentOfThisCommit” (describing the intended purpose of the commit), and “faultLocalization” (an array of JSON objects). The “faultLocalization” array should contain ten JSON objects, each with four fields: “lineNumber” (indicating the line number of the suspicious code), “codeContent” (showing the actual code), “reason” (explaining why this location is identified as potentially faulty) and "score" (a number between 1 and 10 indicating the suspicion level).   

Be concise and deterministic — avoid generic language (“may be wrong”) and point to *specific failure modes*.  Prioritize statement lines over declaration lines.
"""
```

```python
import unicodedata
import re

def sanitize_text(text: str, replace_with: str = "�") -> str:
    """
    Remove ou substitui caracteres Unicode inválidos (surrogates) e normaliza o texto.
    É seguro para logs, JSON e escrita em arquivos UTF-8.

    Args:
        text (str): Texto possivelmente com caracteres inválidos (ex: '\udce4').
        replace_with (str): Caractere de substituição para cada surrogate (default: '�').

    Returns:
        str: Texto limpo e normalizado.
    """
    if not isinstance(text, str):
        return text

    # Substitui explicitamente quaisquer surrogates (U+D800–U+DFFF)
    text = re.sub(r'[\ud800-\udfff]', replace_with, text)

    # Normaliza (remove combinações redundantes, garante forma canônica)
    text = unicodedata.normalize("NFC", text)

    # Re-encode/Decode garante UTF-8 bem-formado
    text = text.encode("utf-8", "replace").decode("utf-8")

    return text
```

## Calculating metrics

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import math

#Adapted (with changes) from JIT-Fine calculation of line-level metrics: https://github.com/jacknichao/JIT-Fine
def get_line_level_metrics(line_score, label):
    scaler = MinMaxScaler()
    line_score = scaler.fit_transform(np.array(line_score).reshape(-1, 1)) 
    pred = np.round(line_score)

    line_df = pd.DataFrame()
    line_df['scr'] = [float(val) for val in list(line_score)]
    line_df['label'] = label
    line_df = line_df.sort_values(by='scr', ascending=False)
    line_df['row'] = np.arange(1, len(line_df) + 1)

    line_df.to_csv('line_df.csv')

    real_buggy_lines = line_df[line_df['label'] == 1]

    top_10_acc = 0
    top_5_acc = 0

    if len(real_buggy_lines) < 1:
        IFA = len(line_df)
        top_20_percent_LOC_recall = 0
        effort_at_20_percent_LOC_recall = math.ceil(0.2 * len(line_df))

    else:
        IFA = line_df[line_df['label'] == 1].iloc[0]['row'] - 1
        label_list = list(line_df['label'])

        all_rows = len(label_list)

        # find top-10 accuracy
        if all_rows < 10:
            top_10_acc = np.sum(label_list[:all_rows]) / len(label_list[:all_rows])
        else:
            top_10_acc = np.sum(label_list[:10]) / len(label_list[:10])

        # find top-5 accuracy
        if all_rows < 5:
            top_5_acc = np.sum(label_list[:all_rows]) / len(label_list[:all_rows])
        else:
            top_5_acc = np.sum(label_list[:5]) / len(label_list[:5])

        #My change: Considers 1 if at least one bug was found
        if top_5_acc > 0:
            top_5_acc = 1.0

        if top_10_acc > 0:
            top_10_acc = 1.0
                
        # find recall
        LOC_20_percent = line_df.head(int(0.2 * len(line_df)))
        buggy_line_num = LOC_20_percent[LOC_20_percent['label'] == 1]
        top_20_percent_LOC_recall = float(len(buggy_line_num)) / float(len(real_buggy_lines))

        # find effort @20% LOC recall
        buggy_20_percent = real_buggy_lines.head(math.ceil(0.2 * len(real_buggy_lines)))
        buggy_20_percent_row_num = buggy_20_percent.iloc[-1]['row']
        effort_at_20_percent_LOC_recall = int(buggy_20_percent_row_num) / float(len(line_df))

    return IFA, top_20_percent_LOC_recall, effort_at_20_percent_LOC_recall, top_10_acc, top_5_acc
```

```python
def find_commit(commit_id):
    projects = file.keys()
    
    for project in projects:
        commits = file[project].keys()
        if commit_id in commits:
            return file[project][commit_id]

    return None
```

```python
import re

def find_original_line(commits, raw_changed_line):
    raw_tokens = raw_changed_line.split()
    added = commits['added']
    added_files = added.keys()
    deleted = commits['deleted']
    deleted_files = deleted.keys()
    best_match = 0
    best_line = None

    def find_in_files(files, best_match, best_line):    
        for file in files:
            lines = added[file]
            for line in lines:
                match = 0
                #line_tokens = line.split()
                line_tokens = re.findall(r"\w+|[^\w\s]", line, re.UNICODE)
                #print('line_tokens = ', line_tokens)
                for token in raw_tokens:
                    if token in line_tokens:
                        match += 1

                if match > best_match:
                    best_match = match
                    best_line = line

        return best_match, best_line

    best_match, best_line = find_in_files(added_files, best_match, best_line)
    best_match, best_line = find_in_files(deleted_files, best_match, best_line)
    
    return best_line
```

```python
def find_in_faults(original_line, faults):
    #print('original_line = ', original_line)
    for i, fault in enumerate(faults):
        code = fault['codeContent']
        
        if original_line and original_line in code:
            #print(code)
            return fault
    return None
```

```python
def escape_codecontent_quotes(text: str) -> str:
    # Escapa aspas duplas dentro dos campos "codeContent"
    def repl(m):
        content = m.group(1)
        # Escapa apenas aspas duplas internas (não escapadas)
        content_fixed = re.sub(r'(?<!\\)"', r'\\"', content)
        return f'"codeContent": "{content_fixed}"'

    return re.sub(r'"codeContent":\s*"([^"]*?)"', repl, text)
```

```python
import re
import json


def parse(raw):
    # 1. extrair intentOfThisCommit
    intent_match = re.search(
        r'"intentOfThisCommit"\s*:\s*"([^"]*)"',
        raw
    )
    intent = intent_match.group(1) if intent_match else None
    
    # 2. extrair a lista faultLocalization como blocos de objetos {...}
    items_raw = re.findall(
        r'\{\s*"lineNumber"[\s\S]*?"score"\s*:\s*\d+\s*\}',
        raw
    )
    
    faults = []
    for item in items_raw:
        # lineNumber
        ln = re.search(r'"lineNumber"\s*:\s*(\d+)', item).group(1)
        ln = int(ln)
    
        # codeContent  (pega tudo depois de "codeContent": " ... " até a aspa antes da vírgula que fecha o campo)
        cc = re.search(r'"codeContent"\s*:\s*"(.*)"\s*,', item, flags=re.S)
        code_content_raw = cc.group(1) if cc else ""
    
        # consertar as aspas internas NÃO escapadas, deixando isso em forma legível
        # regra: dentro do codeContent, toda aspa dupla vira \" se ainda não estiver escapada
        fixed_code_content = []
        escaped = False
        for ch in range(len(code_content_raw)):
            c = code_content_raw[ch]
            if c == '"':
                # verifica se já estava escapada com barra antes
                if ch == 0 or code_content_raw[ch-1] != '\\':
                    fixed_code_content.append('\\"')
                else:
                    fixed_code_content.append('"')  # já era \" no texto
            else:
                fixed_code_content.append(c)
        fixed_code_content = "".join(fixed_code_content)
    
        # reason
        rmatch = re.search(r'"reason"\s*:\s*"(.*)"\s*,\s*"score"', item, flags=re.S)
        reason = rmatch.group(1).strip() if rmatch else ""
    
        # score
        score = int(re.search(r'"score"\s*:\s*(\d+)', item).group(1))
    
        faults.append({
            "lineNumber": ln,
            "codeContent": fixed_code_content,
            "reason": reason,
            "score": score,
        })
    
    result = {
        "intentOfThisCommit": intent,
        "faultLocalization": faults
    }
    return result

```

```python
import json
from tqdm import tqdm

def calculate_metrics(responses, tp_df_test, line_info_df):
    IFA, top_20_percent_LOC_recall, effort_at_20_percent_LOC_recall, top_10_acc, top_5_acc = [], [], [], [], []
        
    for i, response in tqdm(enumerate(responses)):
        commit_id = tp_df_test.loc[i, 'commit_hash']
        line_info = line_info_df[line_info_df['commit_id'] == commit_id]

        if isinstance(response, str):
            try:
                begin = response.index('{')
                end = response.rfind('}')
            except ValueError:
                begin = 0
                end = len(response)
                
            response = response[begin:end+1]

            try:
                cleaned = escape_codecontent_quotes(response)
                response = json.loads(cleaned)
            except json.JSONDecodeError: 
                response = parse(response)
                
        faults = response['faultLocalization']
        if len(faults) == 0:
            continue

        if "score" in faults[0].keys():
            faults = sorted(faults, key=lambda x: x["score"], reverse=True)
            if i == 0:
                print(faults)
            

        commit = find_commit(commit_id)
        for j, row in line_info.iterrows():
            raw_changed_line = row['raw_changed_line']
            original_line = find_original_line(commit, raw_changed_line)
            if not original_line:
                print('NOT FOUND: ', raw_changed_line)
            line_info.loc[j, 'original_line'] = original_line
            line_info['score'] = 0

        for k, row in line_info.iterrows():
            original_line = row['original_line']
            fault = find_in_faults(original_line, faults)
            if fault:
                score = fault['score']
                line_info.loc[k, 'score'] = score

        scores = line_info['score'].tolist()
        labels = line_info['label'].tolist()
        cur_IFA, cur_top_20_percent_LOC_recall, cur_effort_at_20_percent_LOC_recall, cur_top_10_acc, cur_top_5_acc = get_line_level_metrics(scores, labels)

        IFA.append(cur_IFA)
        top_20_percent_LOC_recall.append(cur_top_20_percent_LOC_recall)
        effort_at_20_percent_LOC_recall.append(cur_effort_at_20_percent_LOC_recall)
        top_10_acc.append(cur_top_10_acc)
        top_5_acc.append(cur_top_5_acc)
    return IFA, top_20_percent_LOC_recall, effort_at_20_percent_LOC_recall, top_10_acc, top_5_acc

```

```python
IFA, top_20_percent_LOC_recall, effort_at_20_percent_LOC_recall, top_10_acc, top_5_acc = calculate_metrics(responses, tp_df_test, line_info_df)

print(
        'Top-10-ACC: {:.4f},Top-5-ACC: {:.4f}, Recall20%Effort: {:.4f}, Effort@20%LOC: {:.4f}, IFA: {:.4f}'.format(
            round(np.mean(top_10_acc), 4), round(np.mean(top_5_acc), 4),
            round(np.mean(top_20_percent_LOC_recall), 4),
            round(np.mean(effort_at_20_percent_LOC_recall), 4), round(np.mean(IFA), 4))
    )
```

```python
import json
from tqdm import tqdm

def gen_error_analysis(responses, tp_df_test, line_info_df):
    fps = []
        
    for i, response in tqdm(enumerate(responses)):
        commit_id = tp_df_test.loc[i, 'commit_hash']
        line_info = line_info_df[line_info_df['commit_id'] == commit_id]

        if isinstance(response, str):
            try:
                begin = response.index('{')
                end = response.rfind('}')
            except ValueError:
                begin = 0
                end = len(response)
                
            response = response[begin:end+1]

            try:
                cleaned = escape_codecontent_quotes(response)
                response = json.loads(cleaned)
            except json.JSONDecodeError: 
                response = parse(response)
                
        faults = response['faultLocalization']
        if len(faults) == 0:
            continue

        if "score" in faults[0].keys():
            faults = sorted(faults, key=lambda x: x["score"], reverse=True)
            if i == 0:
                print(faults)

        commit = find_commit(commit_id)
        for j, row in line_info.iterrows():
            raw_changed_line = row['raw_changed_line']
            original_line = find_original_line(commit, raw_changed_line)
            if not original_line:
                print('NOT FOUND: ', raw_changed_line)
            line_info.loc[j, 'original_line'] = original_line
            line_info['score'] = 0

        for k, row in line_info.iterrows():
            original_line = row['original_line']
            fault = find_in_faults(original_line, faults)
            if fault:
                score = fault['score']
                line_info.loc[k, 'score'] = score
                if score > 0 and row['label'] == 0:
                    #false positive
                    fault['project'] = tp_df_test.loc[i, 'project']
                    fps.append(fault)

    fps = sorted(fps, key=lambda x: x["score"], reverse=True)
    return fps
```

```python
import json
from tqdm import tqdm

def gen_false_negatives(responses, tp_df_test, line_info_df):
    fns = []
        
    for i, response in tqdm(enumerate(responses)):
        commit_id = tp_df_test.loc[i, 'commit_hash']
        line_info = line_info_df[line_info_df['commit_id'] == commit_id]

        if isinstance(response, str):
            try:
                begin = response.index('{')
                end = response.rfind('}')
            except ValueError:
                begin = 0
                end = len(response)
                
            response = response[begin:end+1]

            try:
                cleaned = escape_codecontent_quotes(response)
                response = json.loads(cleaned)
            except json.JSONDecodeError: 
                response = parse(response)
                
        faults = response['faultLocalization']
        if len(faults) == 0:
            continue

        if "score" in faults[0].keys():
            faults = sorted(faults, key=lambda x: x["score"], reverse=True)
            if i == 0:
                print(faults)

        commit = find_commit(commit_id)
        for j, row in line_info.iterrows():
            raw_changed_line = row['raw_changed_line']
            original_line = find_original_line(commit, raw_changed_line)
            if not original_line:
                print('NOT FOUND: ', raw_changed_line)
            line_info.loc[j, 'original_line'] = original_line
            line_info['score'] = 0

        for k, row in line_info.iterrows():
            original_line = row['original_line']
            fault = find_in_faults(original_line, faults)
            if row['label'] == 1 and fault is None:
                fns.append(row)

    df_fns = pd.DataFrame(fns)
    return df_fns
```
